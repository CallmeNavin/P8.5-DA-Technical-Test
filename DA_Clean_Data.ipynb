```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
Sales = pd.read_excel('Database - Q3_2020_Sheet Sales.xlsx')
MKT = pd.read_excel('Database - Q3_2020_Sheet MKT.xlsx')
Van_don = pd.read_excel('Database - Q3_2020_Sheet Van đơn.xlsx')

# Completeness
## Chuẩn hóa
Sales = Sales.replace(["", " ", "NA", "N/A", "na", "null", "--"], pd.NA)
MKT = MKT.replace(["", " ", "NA", "N/A", "na", "null", "--"], pd.NA)
Van_don = Van_don.replace(["", " ", "NA", "N/A", "na", "null", "--"], pd.NA)
## Đếm %NA
print("=== Sales ===")
print((Sales.isna().mean()*100).round(2))
print("\n=== MKT ===")
print((MKT.isna().mean()*100).round(2))
print("\n=== Van_don ===")
print((Van_don.isna().mean()*100).round(2))
## Hướng xử lý: Loại bỏ các cột toàn NaN
Sales.drop(columns= ['Sales Admin xác nhận Type of Lead', 'Ngày hẹn gọi lại', 'Tỉnh/TP'], errors= 'ignore', inplace=True)
Van_don.drop(columns= ['Đơn vị tính', 'Serial', 'Ghi chú sản phẩm', 'Lý do hủy đơn', 'Tags đơn hàng', 'Ghi chú đơn hàng', 'Ngày hẹn giao'], errors= 'ignore', inplace=True)

# Validity
## Check Dtype từng cột
print("=== Sales ===")
print(Sales.info())
print("\n=== MKT ===")
print(MKT.info())
print("\n=== Van_don ===")
print(Van_don.info())

# Consistency & Integrity
## So sánh số tiền giữa hai sheet Sales & Van_don
print('Tổng tiền_Sheet Sales: ', Sales['Tổng tiền'].sum())
print('Tổng tiền hàng_Sheet Van_don: ', Van_don['Tổng tiền hàng'].sum())

# Uniqueness
## Check trùng ID mã đơn hàng (Van_don)
print('Số mã đơn hàng trùng trong Van_don', Van_don['Mã đơn hàng'].duplicated().sum())
## Check unique value count
print("Van_don - Tổng dòng:", len(Van_don), "| Unique Mã đơn hàng:", Van_don['Mã đơn hàng'].nunique())
## Hướng xử lý: Group by mã đơn hàng do số mã đơn unique > số dòng
Van_don.groupby("Mã đơn hàng")["Tổng tiền hàng"].sum()

# Timeliness
Sales['Date'] = pd.to_datetime(Sales['Date'], errors='coerce')
print("Sales:", Sales['Date'].min(), "→", Sales['Date'].max())
print("MKT:",   MKT['Date'].min(), "→", MKT['Date'].max())
print("Vận đơn:", Van_don['Ngày giao hàng'].min(), "→", Van_don['Ngày giao hàng'].max())

# Export cleaned data
Sales.to_csv("Sales_cleaned.csv", index=False, encoding="utf-8-sig")
MKT.to_csv("MKT_cleaned.csv", index=False, encoding="utf-8-sig")
Van_don.to_csv("Van_don_cleaned.csv", index=False, encoding="utf-8-sig")
files.download('Van_don_cleaned.csv')
```
